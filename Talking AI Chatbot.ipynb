```python
import speech_recognition as sr
import pyttsx3
import os
import re
from datetime import datetime

# Set your single Eleven Labs API key here
API_KEY = "API_KEY"  # API key for both TTS and STT

# Initialize speech engine for fallback TTS
engine = pyttsx3.init()

# Set speech rate (speed of speech)
engine.setProperty('rate', 150)

# Set volume level
engine.setProperty('volume', 1)

def speak(text):
    """Converts text to speech using Eleven Labs API."""
    url = "https://api.elevenlabs.io/v1/text-to-speech"
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }
    payload = {
        "voice": "en_us_male",  # Use the desired voice model
        "text": text
    }

    response = requests.post(url, headers=headers, json=payload)
    
    if response.status_code == 200:
        audio_content = response.content
        with open("output_audio.mp3", "wb") as audio_file:
            audio_file.write(audio_content)
        os.system("start output_audio.mp3")  
    else:
        engine.say(text)  # Fallback to pyttsx3 if Eleven Labs fails
        engine.runAndWait()

def listen():
    """Listens for a command from the microphone and converts it to text."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for command...")
        audio = recognizer.listen(source)
    
    try:
        command = recognizer.recognize_google(audio)
        print(f"You said: {command}")
        return command
    except sr.UnknownValueError:
        print("Sorry, I didn't catch that.")
        return None
    except sr.RequestError:
        print("There was an error with the speech recognition service.")
        return None

def get_weather():
    """Returns static weather information (no API key needed)."""
    return "The current weather in London is clear sky with a temperature of 20°C."

def get_time():
    """Returns the current time."""
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    return f"The current time is {current_time}."

def get_date():
    """Returns the current date."""
    today = datetime.today()
    current_date = today.strftime("%B %d, %Y")
    return f"Today's date is {current_date}."

def dynamic_response(user_input):
    """Generates a dynamic response based on the user's input."""
    # Convert user input to lowercase for easier matching
    user_input = user_input.lower()

    # Detecting intents based on keywords (No need for hardcoded responses)
    if "weather" in user_input:
        return get_weather()
    
    elif "time" in user_input:
        return get_time()

    elif "date" in user_input:
        return get_date()

    elif "hello" in user_input or "hi" in user_input:
        return "Hello! How can I assist you today?"
    
    elif "how are you" in user_input:
        return "I'm just a program, but I'm doing well. How about you?"

    elif "bye" in user_input or "exit" in user_input:
        return "Goodbye! Have a great day!"

    elif "your name" in user_input:
        return "I am your personal chatbot."

    elif "help" in user_input:
        return "You can ask me about the weather, time, date, or just chat with me."

    else:
        return "Sorry, I didn't understand that. Could you please rephrase?"

def main():
    """Main function for the conversation loop."""
    speak("Hello, how can I assist you today?")
    
    while True:
        # Get input from the user via microphone
        user_input = listen()
        
        if user_input:
            if "exit" in user_input.lower() or "bye" in user_input.lower():
                speak("Goodbye!")
                break
            else:
                # Get chatbot response dynamically based on user input
                response = dynamic_response(user_input)
                speak(response)
        else:
            speak("I didn't catch that. Please say something.")

if __name__ == "__main__":
    main()
```
